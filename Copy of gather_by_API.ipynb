```python
# Generate 100 random dates between 2005 and 2022

import random
import datetime

start_date = datetime.date(2005, 1, 1)
end_date = datetime.date(2022, 12, 31)

random_dates = []
for i in range(100):
    time_between_dates = end_date - start_date
    days_between_dates = time_between_dates.days
    random_number_of_days = random.randrange(days_between_dates)
    random_date = start_date + datetime.timedelta(days=random_number_of_days)
    formatted_date = random_date.strftime("%Y-%m-%d")
    random_dates.append(formatted_date)

random_dates
```

```python
all_data = []
for date in random_dates:
  # Get docket information from the API
  url = f"https://api.regulations.gov/v4/comments?filter[postedDate]={date}&api_key={api_key}"

  response = requests.get(url)
  data = response.json()
  all_data.append(data)
```

```python
all_data
```

```python
ids = []

for data in all_data:
  for comment in data['data']:
    ids.append(comment['id'])

print(ids)
```

```python
comments = []

for id in ids:
  # Get docket information from the API
  url = f"https://api.regulations.gov/v4/comments/{id}?api_key={api_key}"

  response = requests.get(url)
  data = response.json()

  comments.append(data)
```

```python
df = pd.DataFrame([c['data'] for c in comments if 'data' in c])
```

```python
df['comment'] = df['attributes'].apply(lambda x: x['comment'])
```

```python
df['docketId'] = df['attributes'].apply(lambda x: x['docketId'])
```

```python
def get_docketId(id):
  docketId = re.sub(r'-[^-]*$', '', id)
  return docketId

df['docketId'] = df.id.apply(get_docketId)
```

```python
import openai
openai.api_key = ''
```

```python
def ask_chatgpt_comments(review):
  # define prompt for ChatGPT
  p = f'''The following is a comment from a person about a federal regulation: \n\n {review} \n\n'''
  p += '''Does this person support or oppose the regulation? Return 1 if they support, 0 if not. Do not return any other text.''' # 👈👈👈👈👈👈👈👈👈👈👈👈👈👈👈👈👈

  # generate the response
  response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=p,
        temperature=0,
        max_tokens=500,
        top_p=1,
        frequency_penalty=0.5,
        presence_penalty=0,
        stop=["12."]
      )

  # grab our text from the repsonse
  text = response['choices'][0]['text']
  return text.strip()
```

```python
golden_set_df['gpt_answer'] = golden_set_df.comment.apply(ask_chatgpt_comments)
```
